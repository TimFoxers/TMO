## 1. Текстовое описание набора данных
В лабораторной работе используется датасет "2021 World Population".

Имеются следующие данные:
1. ISO код страны (iso_code)
2. Страна (country)
3. Обновленные данные за 2021-й год о населении (2021_last_updated)
4. Данные за 2020-й год о населении (2020_population)
5. Площадь страны (area)
6. Плотность населения на квадратный километр (density_sq_km)
7. Рост населения (growth_rate)
8. Процент населения относительно мира (world_%)
9. Номер по рейтингу (rank)

Загрузим выбранный датасет

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="ticks")
```

```python
data = pd.read_csv('2021_population.csv', sep = ",")
```

Преобразуем цифровые значения к рабочему виду:

```python
for i in range(len(data['2020_population'])):
    data['2020_population'][i] = data['2020_population'][i].replace(",", "")        
    data['2020_population'][i] = str(int(float(data['2020_population'][i])))
```

```python
data['2020_population'] = data['2020_population'].astype(str).astype(int)
```

```python
for i in range(len(data['2021_last_updated'])):
    data['2021_last_updated'][i] = data['2021_last_updated'][i].replace(",", "")        
    data['2021_last_updated'][i] = str(int(float(data['2021_last_updated'][i])))
```

```python
data['2021_last_updated'] = data['2021_last_updated'].astype(str).astype(int)
```

```python
for i in range(len(data['area'])):
    data['area'][i] = data['area'][i].replace(",", "")
    data['area'][i] = data['area'][i].replace(" sq_km", "")   
    data['area'][i] = str(int(float(data['area'][i])))
```

```python
data['area'] = data['area'].astype(str).astype(int)
```

```python
for i in range(len(data['density_sq_km'])):
    data['density_sq_km'][i] = data['density_sq_km'][i].replace(",", "")
    data['density_sq_km'][i] = data['density_sq_km'][i].replace("/sq_km", "")   
    data['density_sq_km'][i] = str(int(float(data['density_sq_km'][i])))
```

```python
data['density_sq_km'] = data['density_sq_km'].astype(str).astype(int)
```

```python
for i in range(len(data['growth_rate'])):
    data['growth_rate'][i] = data['growth_rate'][i].replace("%", "")   
    data['growth_rate'][i] = str(float(data['growth_rate'][i]))
    
data['growth_rate'] = data['growth_rate'].astype(str).astype(float)
```

```python
for i in range(len(data['world_%'])):
    data['world_%'][i] = data['world_%'][i].replace("%", "")   
    data['world_%'][i] = str(float(data['world_%'][i]))
    
data['world_%'] = data['world_%'].astype(str).astype(float)
```
